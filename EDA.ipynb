9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.
- The duration of tracks varies noticeably across artists — some artists produce consistently longer tracks, while others prefer shorter ones.
 - A box or violin plot shows this variation clearly, with artists like Drake and Eminem having wider duration ranges compared to others.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/spotify.csv").drop_duplicates()
plt.figure(figsize=(10,5))
sns.boxplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=90)
plt.title("Variation of Track Duration Across Artists")
plt.show()
```

10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot.
 - The below violin plot shows that popularity distributions differ by artist, with some having tightly clustered high scores (consistent hits) and others showing wider variation.

```python
#10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/spotify.csv").drop_duplicates()
plt.figure(figsize=(10,5))
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=90)
plt.title("Distribution of Track Popularity Across Artists")
plt.show()
```
